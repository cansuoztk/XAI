```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn import datasets

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn import datasets
!pip install interpret
import interpret
from interpret.glassbox import ExplainableBoostingClassifier
from interpret import show
```

```python
df=pd.read_csv("Iris.csv")

df.head()
```

```python
df = df.rename(columns={"SepalLengthCm":"Sepal_length_cm", "SepalWidthCm":"Sepal_width_cm","PetalLengthCm":"Petal_length_cm","PetalWidthCm":"Petal_width_cm"  })
df
```

```python
Caracte = ["Sepal_length_cm","Sepal_width_cm","Petal_length_cm","Petal_width_cm" ]
obje = df.Species

X = df[Caracte]
y = obje
```

verilerimizi test ve eğitim kümesine ayırıyoruz.

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.33,random_state = 42)
```

Açıklanabilir Artırma Makinesi (EBM), otomatik etkileşim algılamalı, ağaç tabanlı, döngüsel bir gradyan artırıcı Genelleştirilmiş Katkı Modelidir. EBM'ler tamamen yorumlanabilir kalırken genellikle son teknoloji kara kutu modelleri kadar doğrudur. EBM'lerin eğitilmesi genellikle diğer modern algoritmalara göre daha yavaş olsa da, EBM'ler son derece kompakt ve tahmin süresinde hızlıdır.

```python
ebm = ExplainableBoostingClassifier()
ebm.fit(x_train, y_train)
```

```python
ebm_global = ebm.explain_global()
show(ebm_global)
```

```python
ebm_local = ebm.explain_local(x_test, y_test)
show(ebm_local)
```
